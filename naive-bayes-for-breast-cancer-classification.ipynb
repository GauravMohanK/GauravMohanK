# Naive Bayes Classifier: 97% Accuracy 

I have been obssessively looking for a more compact model than my PCA aided Logistic Regression for this project. My little research led me to this one. It also has better accuracy.

```python
import numpy as np
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import LinearRegression
from sklearn.metrics import accuracy_score
```

```python
data = load_breast_cancer()
label_names = data['target_names']
labels = data['target']
feature_names = data['feature_names']
feature = data['data']
```

```python
print(label_names)
print(np.array(np.unique(labels, return_counts = True)))
print(feature_names)
```

```python
print(feature[0])
```

```python
train, test, train_labels, test_labels = train_test_split(feature, labels, test_size = 0.2, random_state = 42)
```

```python
gnb = GaussianNB()
gnb.fit(train, train_labels)
preds = gnb.predict(test)
print(preds, '\n')
```

```python
print(accuracy_score(test_labels, preds))
```

97% is the accuracy!!
