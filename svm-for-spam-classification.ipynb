# Spam Classification With Support Vector Machine 

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn import svm
```

```python
spam = pd.read_csv('/kaggle/input/spam-dataset/spam_ham_dataset.csv')
```

Here we have imported all the necessary libraries to build our SVM model

We assign variable 'spam' to our data and read the data in pandas dataframe object

```python
spam.head()
```

We apply .head() method to view top 5 rows and columns of our dataframe

```python
z = spam["text"]
y = spam["label"]

z_train, z_test, y_train, y_test = train_test_split(z, y, test_size= 0.2)
```

Here we assign 'text' column vector and 'label' column vector to z and y axes respectively and then we apply train_test_split function from sklearn package to split 20% of our data into test data (indicated by test_size = 0.2), that will be used to gauge accuracy of the model later

```python
cv = CountVectorizer()
features = cv.fit_transform(z_train)
```

CountVectorizer() is being used here to assign random numbers to our texts

```python
print(features)
```

By using print() function I can see vectorized or tokenized texts and their frequency

```python
model = svm.SVC()
model.fit(features, y_train)
```

Here we assign svm.SVC class to our model and then train it by feeding it our new vectorized features against label 

```python
features_test = cv.transform(z_test)
print("Accuracy: {}".format(model.score(features_test, y_test)))
```

Finally, we calculate accuracy score by transforming our test data and running the model feeding this data. 97% is the accuracy.
